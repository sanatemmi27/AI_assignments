# ONE HOT ENCODING

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse= False)

encode = ohe.fit_transform(df[["SEX", "MARRIAGE", "EDUCATION"]])
```

```python
df1 = pd.DataFrame(encode, columns= ["Male", "Female", "Married", "Single", "OtherMarriage", "GraduateSchool", "University", "HighSchool", "OtherSchool"])
df1 = df1.reset_index(drop=True)
```

```python
df= df.drop(["SEX","EDUCATION", "MARRIAGE"],1)
```

```python
data = pd.concat([df, df1], axis=1)
```

```python
data.head()
```

# SCALING

```python
cols = data.columns
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()

scaled = scalar.fit_transform(data)
data = pd.DataFrame(scaled, columns=cols)
```

```python
X= data.drop(["ID","target"],1)
y=data["target"]
```

# MODELLING

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

# ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# first hidden layer
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
# units:inputs nodes that are going to add
classifier.add(Dense(units = 16, activation = 'relu', input_dim = 29))
```

```python
# Adding the second hidden layer
# units: Positive integer, dimensionality of the output space.
classifier.add(Dense(units = 16, activation = 'relu'))
```

```python
# Adding the output layer
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 30)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

# FINAL REPORT

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
