```python
# to remove the "?" values acting as missing values
def clean_numerical(col_name):
    m=df[(df[col_name]!="?")][col_name].astype(float).mean()
    df[col_name] = df[col_name].str.replace("?",str(m)).astype(float)
    return df
```
