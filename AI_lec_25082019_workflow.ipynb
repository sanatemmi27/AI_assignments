# Modelling

```python
X = df.drop(["ID","default.payment.next.month"],1)
y = df["default.payment.next.month"]
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test = tts(X,y,random_state =40,test_size=0.30)
from sklearn.linear_model import LogisticRegression
```

```python
# Logistic Regression 

from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
lr = LogisticRegression(class_weight="balanced") 
lr.fit(X_train, y_train) 
y_pred = lr.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
from datetime import datetime
pre_log = precision_score(y_pred,y_test)
recall_log = recall_score(y_pred,y_test)
time_log = str(datetime.now())
LR_score = accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
#-----------> Random Forest with GRID SEARCH
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
from sklearn.model_selection import GridSearchCV

param_grid = [{"n_estimators":[3,10,30], "max_features":[2,4]} #max features not greater than no.of features in dataset
              ,{"bootstrap": [False],"n_estimators":[3,10],"max_features":[2,3,4]},]
rfc = GridSearchCV(rfc,param_grid,cv=5,scoring="accuracy")
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
pre_RF = precision_score(y_pred,y_test)
recall_RF = recall_score(y_pred,y_test)
time_RF = str(datetime.now())
RF_score = accuracy_score(y_pred,y_test)
```

# Results

```python
Model = ["Logistic Regression","Random Forest Classifier"]
#cross_val= [crossval_log,crossval_rf,crossval_knn,crossval_svc]
#matrix = [log_matrix,rf_matrix,knn_matrix,svc_matrix]
precision = [pre_log,pre_RF]
recall = [recall_log,recall_RF]
time = [time_log,time_RF]
score = [LR_score, RF_score]
report=pd.DataFrame({'TimeStamp':time,'Model': Model,'Precision': precision,'Recall': recall, "Accuracy": score})
report
```
